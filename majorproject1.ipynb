```python
#create datframe
import pandas as pd
df= pd.read_csv('/content/Reddit_Data.csv')
df
```

```python
df=df[['clean_comment','category']]
```

```python
df['clean_comment']=df['clean_comment'].map({"negative": 0, "neutral": 1, "positive": 2})
```

```python
df['clean_comment']=df['clean_comment'].astype(str)
```

```python
#split the data into training and testing sets
```

```python
x=df['clean_comment']
y=df['category']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#preprocessing
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(stop_words="english")
x_train_transformed = vectorizer.fit_transform(x_train.astype(str)).toarray()
x_test_transformed = vectorizer.transform(x_test.astype(str)).toarray()
```

```python
#apply classifier
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
```

```python
#fit the model
lr_model.fit(x_train_transformed,y_train)
```

```python
#predict the output
y_pred = lr_model.predict(x_test_transformed)
y_pred
```

```python
#accuracy
from sklearn.metrics import accuracy_score
lr_accuracy = accuracy_score(y_test, y_pred)*100
lr_accuracy
```

```python
#for rest of the classifiers and also logistic
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
classifiers ={
    "Logistic Regression": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier()
}
```

```python
for name, classifier in classifiers.items():
    classifier.fit(x_train_transformed, y_train)
    y_pred = classifier.predict(x_test_transformed)
    accuracy = accuracy_score(y_test, y_pred)*100
    print(f"{name} Accuracy: {accuracy}")
```
